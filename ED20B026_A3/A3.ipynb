```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the system of differential equations
def hopf_oscillators(y, t, mu, A, omega1, omega2, theta):
    r1, r2, phi1, phi2 = y

    dydt = [
        (mu - r1**2) * r1 + A * r2 * np.cos(phi2 - phi1 + theta),
        (mu - r2**2) * r2 + A * r1 * np.cos(phi1 - phi2 - theta),
        omega1 + A * (r2 / r1) * np.sin(phi2 - phi1 + theta),
        omega2 + A * (r1 / r2) * np.sin(phi1 - phi2 - theta)
    ]

    return dydt

# Set parameters
mu = 0.1  # Adjust as needed
omega1 = 5
omega2 = 5
A = 0.5  # Initial guess for A, adjust as needed
theta = np.radians(98)  # Convert degrees to radians

# Set initial conditions
y0 = [1, 1, np.radians(220), np.radians(47)]  # You may need to adjust initial conditions based on your requirements

# Set time points for integration
t = np.linspace(0, 20, 1000)  # Adjust the time span and number of points as needed

# Solve the system of differential equations
solution = odeint(hopf_oscillators, y0, t, args=(mu, A, omega1, omega2, theta))

# Extract theta1 and theta2 values
theta1_values =solution[:, 2]
theta2_values =solution[:, 3]
f=20
# Plot the results
plt.figure(figsize=(12, 6))
# plt.plot(t, np.degrees(theta1_values), label=r'$\theta_1$')
# plt.plot(t, np.degrees(theta2_values), label=r'$\theta_2$')
plt.plot(t, np.degrees(solution[:, 2] - solution[:, 3]), label='Phase Difference')
plt.axhline(y=98, color='r', linestyle='--', label='Desired Phase Difference = 98 degrees')
plt.title("Phase Difference vs time",fontsize=f)
plt.xlabel('Time',fontsize=f)
plt.ylabel(r'$\theta_2-\theta_1$ (degrees)',fontsize = f)
plt.xticks(fontsize=f-4)
plt.yticks(fontsize=f-4)
plt.legend(fontsize=f)
plt.grid(True)

plt.show()
```
